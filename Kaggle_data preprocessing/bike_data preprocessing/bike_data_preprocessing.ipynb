```python
data = pd.read_csv('/content/Drive/MyDrive/논문/bike_data 전처리/bike.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['cnt'] = (data['cnt'] - data['cnt'].min())/(data['cnt'].max() - data['cnt'].min()) # 정규화
```

```python
data = data.iloc[:, 1:] # 첫번째 열 제거
```

```python
data_re = data.sort_values(['dteday', 'hr']) # 연속적인 시계열 데이터로 변환 (시계열 퍼지화를 위해)
```
